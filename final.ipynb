```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as smf

%matplotlib inline

from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
df = pd.read_csv('Kickstartercombo.csv')
df.head()
```

```python
df.columns
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
df.head()
```

```python
df.shape
```

```python
df.drop("name", axis=1, inplace=True)
```

```python
df.dtypes
```

```python
fordf = df.loc[df['currency'] != 'USD'] 
fordf.shape
```

```python
print(df.isnull().sum())
fordf.isnull().sum()
```

```python
fordf["usd_goal_real"].fillna(fordf["goal"], inplace=True)
fordf["usd_pledged_real"].fillna(fordf["pledged"], inplace=True)
print(fordf.isnull().sum())

print(df.isnull().sum())
df["usd_goal_real"].fillna(df["goal"], inplace=True)
df["usd_pledged_real"].fillna(df["pledged"], inplace=True)
print(df.isnull().sum())
```

```python
fordf = fordf.drop(['goal','pledged','usd pledged'], axis=1)
fordf.shape
```

```python
df = df.drop(['goal','pledged','usd pledged'], axis=1)
df.shape
```

```python
df.rename(columns={'usd_goal_real': 'goal','usd_pledged_real':'pledged'}, inplace =True)
df.columns
```

```python
fordf.rename(columns={'usd_goal_real': 'goal','usd_pledged_real':'pledged'}, inplace =True)
```

```python
print(fordf.columns)
df.columns
```

```python
df.loc[df['currency'] != 'USD', 'goal'] = fordf['goal'] 
df.loc[df['currency'] != 'USD', 'pledged'] = fordf['pledged'] 
df.shape
```

```python
df = df.drop('currency', axis=1)
```

```python
df = df.drop('ID', axis=1)
df
```

```python
df.isnull().sum()
```

```python
df.loc[(df['state'] != 'successful') & (df['state'] != 'live'), 'state']  = 0
df.head()
```

```python
df.loc[df['state'] != 0, 'state'] = 1
df.head()
```

```python
df.dtypes
```

```python
df['launched']=pd.to_datetime(df['launched'])
df['deadline']=pd.to_datetime(df['deadline'])
df.dtypes
```

```python
df['year'], df['month'] = df['launched'].dt.year, df['launched'].dt.month
df
```

```python
df = df.drop('deadline', axis=1)
df
```

```python
df.loc[df['country'] == 'US', 'country'] = 1
df
```

```python
df.loc[df['country'] != 1, 'country'] = 0
df
```

```python
maincat_df = df.copy()
```

```python
maincat_df = maincat_df.drop('category', axis=1)
```

```python
maincat_df = maincat_df.drop('launched', axis=1)
```

```python
maincat_df = pd.get_dummies(data=maincat_df, columns=['main_category'])
maincat_df.head()
```

```python
y = maincat_df['state']
```

```python
X = maincat_df.drop('state', axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=15)
```

```python
tree = tree.DecisionTreeClassifier()
```

```python
tree.fit(X_train, y_train)
```

```python
tree.score(X_train, y_train)
```

```python
y_predict = tree.predict(X_test)

tree.score(X_test, y_test)
```

```python
pd.DataFrame(
    confusion_matrix(y_test, y_predict),
    columns=['Predicted Failed', 'Predicted Successful'],
    index=['Actual Failed', 'Actual Successful']
)
```

```python
print(classification_report(y_test, y_predict))
```

```python
maincat_df2 = maincat_df.copy()
maincat_df2 = maincat_df2.drop('backers', axis=1)
maincat_df2 = maincat_df2.drop('pledged', axis=1)

y2 = maincat_df2['state']
X2 = maincat_df2.drop('state', axis=1)
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.3, random_state=15)
tree2 = tree.DecisionTreeClassifier()
tree2.fit(X2_train, y2_train)
```

```python
tree2.score(X2_train, y2_train)
```

```python
y2_predict = tree2.predict(X2_test)

tree2.score(X2_test, y2_test)
```

```python
pd.DataFrame(
    confusion_matrix(y2_test, y2_predict),
    columns=['Predicted Failed', 'Predicted Successful'],
    index=['Actual Failed', 'Actual Successful']
)
```

```python
print(classification_report(y2_test, y2_predict))
```
